```python
import cv2 as cv    
import numpy as np
import matplotlib.pyplot as plt


im= cv.imread("im01small.png", cv.IMREAD_GRAYSCALE)
assert im is not None, "Image not found"


scale = 2.5

rows = int(scale * im.shape[0])
cols = int(scale * im.shape[1])

zoomed_image = np.zeros((rows, cols), dtype=im.dtype)
for i in range(0,rows):
    for j in range(0,cols):
        zoomed_image[i, j] = im[int(i / scale), int(j / scale)]

# Display the images

fig, ax= plt.subplots(1,2)
ax[0].imshow(im, cmap='gray')
ax[0].set_title('Original Image')
ax[1].imshow(zoomed_image, cmap='gray')
ax[1].set_title('Zoomed Image')

plt.show()

# Calculate normalized sum of squared differences (SSD)
#ssd_nn = np.sum((zoomed_image.astype("int") - im.astype("int")) ** 2)

#print("SSD (Nearest Neighbor):", ssd_nn)
```
