```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

im = cv.imread('im01.png', cv.IMREAD_GRAYSCALE)
#imgc = cv.cvtColor(im, cv.COLOR_BGR2RGB)

scale =2.5
rows = int(im.shape[0] * scale)
cols = int(im.shape[1] * scale)

zommed = np.zeros((rows, cols), dtype=np.uint8)
for i in range(0,rows):
    for j in range(0,cols):
        zommed[i, j] = im[int(i/scale), int(j/scale)]


fig, ax= plt.subplots(1,2)
ax[0].imshow(im)
ax[0].set_title('Original')
ax[1].imshow(zommed)
ax[1].set_title('zoomed')

plt.show()


```
