Intensity Transformation : Identity Transformation Code 1

```python
%matplotlib inline
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
transform = np.arange(0,256).astype('uint8')
fig,ax = plt.subplots()
ax.plot (transform)
ax.set_xlabel('input ,$f(\mathbf{x})$')
ax.set_ylabel('output,$\\mathrm{T}[f(\mathbf{x})]$')
ax.set_xlim(0,255)
ax.set_ylim(0,255)
ax.set_aspect('equal')
plt.savefig('transform identity.png')
plt.show()
img_orig = cv.imread('emma.jpg',cv.IMREAD_GRAYSCALE)
print(img_orig.shape)
cv.namedWindow('image',cv.WINDOW_AUTOSIZE)
cv.imshow('image',img_orig)
cv.waitKey(0)
image_transformed = cv.LUT(img_orig,transform)
cv.imshow('image',image_transformed)
```

Identity Transform Using Numpy Indexing

```python
# identity transform using numpy

%matplotlib inline
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

transform = np.arange(0,256).astype('uint8')
fig,ax = plt.subplots()
ax.plot (transform)
ax.set_xlabel('input ,$f(\mathbf{x})$')
ax.set_ylabel('output,$\\mathrm{T}[f(\mathbf{x})]$')
ax.set_xlim(0,255)
ax.set_ylim(0,255)
ax.set_aspect('equal')
plt.show()

img_orig = cv.imread('emma.jpg',cv.IMREAD_GRAYSCALE)
print(img_orig.shape)
cv.namedWindow('image',cv.WINDOW_AUTOSIZE)
cv.imshow('image',img_orig)
cv.waitKey(0)
image_transformed = transform[img_orig]
cv.imshow('image',image_transformed)
cv.waitKey(0)
cv.destroyAllWindows()
```

Intensity Transform : Intensity Windowing 3

```python
#Intensity Windowing

%matplotlib inline
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt

c= np.array([(100,50),(150,200)])

t1 = np.linspace(0,c[0,1],c[0,0]+1-0).astype('uint8')
print(len(t1))
t2 = np.linspace(c[0,1]+1,c[1,1],c[1,0]-c[0,0]).astype('uint8')
print(len(t2))  
t3 = np.linspace(c[1,1]+1,255,255-c[1,0]).astype('uint8')
print(len(t3))

transform = np.concatenate((t1,t2),axis=0).astype('uint8')
transform = np.concatenate((transform,t3),axis=0).astype('uint8')
print(len(transform))

fig,ax = plt.subplots()
ax.plot(transform)
ax.set_xlabel('input ,$f(\mathbf{x})$')
ax.set_ylabel('output,$\\mathrm{T}[f(\mathbf{x})]$')
ax.set_xlim(0,255)
ax.set_ylim(0,255)
ax.set_aspect('equal')
plt.show()

img_orig = cv.imread('emma.jpg',cv.IMREAD_GRAYSCALE)
cv.namedWindow('image',cv.WINDOW_AUTOSIZE)
cv.imshow('image',img_orig)
cv.waitKey(0)
image_transformed = cv.LUT(img_orig,transform)
cv.imshow('image',image_transformed)
cv.waitKey(0)
cv.destroyAllWindows()

```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

fig,ax = plt.subplots()
gamma = [0.2,0.5,1,2,5] 
for g in gamma:
    table = np.array([((i/255.0)**g)*255 for i in np.arange(0,256)]).astype('uint8')
    ax.plot(table,label='$\\gamma ={g}$'.format(g=g))
ax.legend(loc = 'best')
ax.set_xlim(0,255)
ax.set_ylim(0,255)
ax.set_aspect('equal')
ax.set_xlabel(r'input ,$f(\mathbf{x})$')
ax.set_ylabel(r'output,$\mathrm{T}[f(\mathbf{x})]$')
plt.show()
```
