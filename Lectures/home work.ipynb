```python
import numpy as np
import matplotlib.pyplot as plt

m = 2 # Line equation y = m*x + c. m is the slope . c is the y-intercept
c = 1 
x = np.arange (1,10,1) # x values
n = 2*np.random.randn(len(x)) # Random noise
o = np.zeros(x.shape) # y = 0 line
#o[-1] = 20 # outlier 
y = m*x + c + n + o # y values

X = np.concatenate([x.reshape(len(x),1),np.ones((len(x),1))],axis=1) # x values with a column of ones
print(X.transpose()@X) # x transpose times x
B = np.linalg.pinv(X.T@X)@X.T@y # (x transpose times x)^-1 times x transpose times y
print(B) # B = [m,c]
mstar = B[0] # m*
cstar = B[1] # c*

plt.plot([x[0], x[-1]], [m*x[0]+c ,m*x[-1]+c], color='b', linestyle ='-' , linewidth = 2 , label = r'original') # Plot the line
plt.plot([x[0], x[-1]], [mstar*x[0]+cstar ,mstar*x[-1]+cstar], color='r', linestyle ='-' , linewidth = 2 , label = r'Estimated') # Plot the line
plt.plot(x,y,'o', label = 'Noisy  points') # Plot the data

plt.xlabel('x') # x-axis label
plt.ylabel('y') # y-axis label
plt.legend(loc = 'best') # Show the legend
```
